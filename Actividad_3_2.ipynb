```python
data20= pd.read_csv('Gastos y costos 20_sn.csv', index_col=0) 
data21= pd.read_csv('Gastos y costos 21_sn.csv', index_col=0) 
data22= pd.read_csv('Gastos y costos 22_sn.csv', index_col=0) 
data23= pd.read_csv('Gastos y costos 23_sn.csv', index_col=0) 
```

# **2022**

```python
data20.head()
```

```python
data20.info()
```

```python
data20.columns
```

```python
data20_n = data20.drop(['FECHA', 'FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'GASTO', 'DESCRIPCION', 'TIPO', 'STATUS', 'POLIZA'], axis=1)
data20_n.head()
```

```python
data20_o = data20.drop(['TC', 'IMPORTE', 'IVA', 'RET ISR', 'RET IVA', 'TOTAL MX', 'TOTAL SAT', 'Columna1'], axis=1)
data20_o.head()
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data20.plot(kind='box', vert=False)
plt.title("Valores Atípicos del 2020")
plt.show() #dibujamos el diagrama
```

**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data20_n
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido\n", Limite_Superior)
print("Limite inferior permitido\n", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers20A= data20_n[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers20A
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data20_2A= data20_n[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data20_2A
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data20_2A.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean_20A=data20_2A.copy()
data_clean_20A=data_clean_20A.fillna(round(data20_2A.mean(),1))
data_clean_20A
```

```python
data_clean_20A = pd.concat([data20_o, data_clean_20A], axis=1)
data_clean_20A
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean_20A.isnull().sum()
valores_nulos
```

**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data20_n

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido\n", Limite_Superior_iqr)
print("Limite inferior permitido\n", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers20B= data20_n[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers20B
```

```python
#Obtenemos datos limpios del Dataframe
data20_2B= data20_n[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data20_2B
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data20_2B.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean_20B=data20_2B.copy()
data_clean_20B=data_clean_20B.fillna(round(data20_2B.mean(),1))
data_clean_20B
```

```python
# Unimos la columna 'indice_tiempo' con el dataframe

data_clean_20B = pd.concat([data20_o, data_clean_20B], axis=1)
data_clean_20B
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean_20B.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data_clean_20B.plot(kind='box', vert=False)
plt.title("Valores Atípicos del 2020")
plt.show() #dibujamos el diagrama
```
